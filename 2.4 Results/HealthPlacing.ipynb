# Health Index Construction

## 1. Data Operations

I = 0.33 * Accessibility + 0.33 * Offer Quality + 0.33 * Population's Health Quality


- Accessibility: Number of health services (min-max scaling normalization)

- Offer Quality --> avg between: Centers' Status (min-max using count of distinct words in string), Centers' Type (min-max using count of distinct words in string), Centers' Characteristics (min-max using count of distinct words in string)

### ACCESSIBILITY

```python
# Accessibility
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
access_col = 'Number of Services'

# Reshape the data since the scaler expects a 2D array
health_idx_info_df[[access_col]] = scaler.fit_transform(health_idx_info_df[[access_col]])

display(health_idx_info_df)
```

```python
construct_df = pd.DataFrame(health_idx_info_df['Number of Services'])
construct_df.rename(columns={'Number of Services': 'Accessibility'}, inplace=True)
display(construct_df)
```

### OFFER QUALITY

```python
health_idx_info_df.head(5)
```

```python
display(health_idx_info_df[health_idx_info_df['Centers Type'].isnull()])
```

```python
# Count distinct values of centers type string (separated by comma)
def count_distinct_values(row): 
    if row == '':
        return 0
    values = row.split(',')
    distinct_values = set(values)
    return len(distinct_values)

# Apply to Centers Type column
construct_df['Count Centers Type'] = health_idx_info_df['Centers Type'].apply(count_distinct_values)
```

```python
# Apply Min-Max

# Offer Quality
offer_col = 'Count Centers Type'

# Reshape the data since the scaler expects a 2D array
construct_df[[offer_col]] = scaler.fit_transform(construct_df[[offer_col]])
construct_df.rename(columns={'Count Centers Type': 'Offer Quality'}, inplace=True)
display(construct_df)
```

### Population's Health Quality

```python
health_idx_info_df.head()
```

```python
display(health_idx_info_df[health_idx_info_df['People with Overweight or Obesity (%)'].isnull()])
```

```python
health_pop_df = health_idx_info_df.copy()
health_pop_df.drop(columns={'Number of Services', 'Distances', 'Centers Type', 'Territorial Socioeconomic Index (IST)',
                           'Avg. Gini Index (%)', 'Population Density (hab/ha)', 'Net Population Density (hab/residential ha)'}, inplace=True)
health_pop_df['People with Overweight or Obesity (%)'][73] = health_pop_df['People with Overweight or Obesity (%)'][72]
health_pop_df.head()
```

```python
construct_df['Medical Conditions'] = (health_pop_df['Notifiable Diseases Rate']/100 + health_pop_df['People with 4 or more Chronic Disorders (%)']/100 + health_pop_df['People with Overweight or Obesity (%)']/100 + health_pop_df['Low Weight Births (%)']/100 + health_pop_df['People at risk of Poor Mental Health (%)']/100 + health_pop_df['Habitual Smokers (%)']/100)/6
construct_df['Problematic Drug Use Index (ICPD)'] = health_pop_df['Problematic Drug Use Index (ICPD)']
construct_df['Mortality'] = health_pop_df['Population Children Deaths Rate']*3 + health_pop_df['Population Premature Deaths (<70) Rate']*2 + health_pop_df['Population Deaths Rate']*1
construct_df['Age vs Expectancy'] = health_pop_df['Avg. Population Age']/health_pop_df['Avg. Population Life Expectancy']
display(construct_df)
```

```python
# Apply minmax
med_col = 'Medical Conditions'
drug_col = 'Problematic Drug Use Index (ICPD)'
mort_col = 'Mortality'

# Reshape the data since the scaler expects a 2D array
construct_df[[med_col]] = scaler.fit_transform(construct_df[[med_col]])
construct_df[[drug_col]] = scaler.fit_transform(construct_df[[drug_col]])
construct_df[[mort_col]] = scaler.fit_transform(construct_df[[mort_col]])
display(construct_df)
```

```python
construct_df['Population Health'] = (construct_df['Medical Conditions'] + construct_df['Mortality'] + construct_df['Age vs Expectancy'] + construct_df['Problematic Drug Use Index (ICPD)'])/4
construct_df.drop(columns={'Medical Conditions', 'Mortality', 'Age vs Expectancy', 'Problematic Drug Use Index (ICPD)'}, inplace=True)
display(construct_df)
```

```python
construct_df['Health Index'] = (construct_df['Accessibility'] + construct_df['Offer Quality'] - construct_df['Population Health'])/3
display(construct_df)
```

```python
construct_df['Neighborhood Name'] = health_idx_info_df['Neighborhood Name']
display(construct_df)
```

```python
# Top 7
print("7 Neighbourhoods with Highest Health Index:\n")
display(construct_df.nlargest(7, 'Health Index')[['Neighborhood Name', 'Health Index']])


# Low 7
print("7 Neighbourhoods with Lowest Health Index:\n")
display(construct_df.nsmallest(7, 'Health Index')[['Neighborhood Name', 'Health Index']])
```
