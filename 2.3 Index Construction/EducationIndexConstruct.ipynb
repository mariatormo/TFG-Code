### Min-Max Scaling Normalization for number of services

### ACCESSIBILITY

```python
# Accessibility
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
access_col = 'Number of Services'

# Reshape the data since the scaler expects a 2D array
educ_idx_info_df[[access_col]] = scaler.fit_transform(educ_idx_info_df[[access_col]])

display(educ_idx_info_df)
```

```python
construct_df = pd.DataFrame(educ_idx_info_df['Number of Services'])
construct_df.rename(columns={'Number of Services': 'Accessibility'}, inplace=True)
display(construct_df)
```

### OFFER QUALITY

```python
educ_idx_info_df.head(5)
```

```python
# Count distinct values of centers type string (separated by comma)
def count_distinct_values(row): 
    values = row.split(',')
    distinct_values = set(values)
    return len(distinct_values)

def count_religious(row):
    values = row.split('and')
    distinct_values = set(values)
    return len(distinct_values)

# Apply to Centers Type column
construct_df['Count Centers Type'] = educ_idx_info_df['Centers Type'].apply(count_distinct_values)
construct_df['Count Centers Simple Type'] = educ_idx_info_df['Centers Simple Type'].apply(count_distinct_values)
construct_df['Count School Status'] = educ_idx_info_df['School Status'].apply(count_distinct_values)
construct_df['Count Centers Characteristics'] = educ_idx_info_df['Centers Characteristics'].apply(count_religious)
```

```python
#display(construct_df)
```

```python
# Apply Min-Max

# Offer Quality
offer_col_type = 'Count Centers Type'
offer_col_type_2 = 'Count Centers Simple Type'
status_col = 'Count School Status'
charact_col = 'Count Centers Characteristics'

# Reshape the data since the scaler expects a 2D array
construct_df[[offer_col_type]] = scaler.fit_transform(construct_df[[offer_col_type]])
construct_df[[offer_col_type_2]] = scaler.fit_transform(construct_df[[offer_col_type_2]])
construct_df[[status_col]] = scaler.fit_transform(construct_df[[status_col]])
construct_df[charact_col] = construct_df[charact_col]/construct_df[charact_col].max() 
display(construct_df)
```

```python
# Weight Offer Quality (highest values - most variate offer)
construct_df['Offer Quality'] = (construct_df['Count Centers Type'] + construct_df['Count School Status'] + construct_df['Count Centers Characteristics'])/3
construct_df.drop(columns={'Count Centers Type', 'Count Centers Simple Type', 'Count School Status', 'Count Centers Characteristics'}, inplace=True)
display(construct_df)
```

### Population's Education Quality

```python
educ_index_df = educ_idx_info_df.copy()

educ_index_df.drop(columns={'Number of Services', 'Distances', 'Density over Places Ratio', 'School Status', 'Centers Type', 'Total Number of Students',
                           'Centers Simple Type', 'Centers Characteristics', 'Number of Teachers', 'Number of Students', 
                            'Number of Places', 'Avg. Students per Teacher Ratio', 'Territorial Socioeconomic Index (IST)',
                           'Avg. Gini Index (%)', 'Population Density (hab/ha)', 'Net Population Density (hab/residential ha)'}, inplace=True)
display(educ_index_df)
```

```python
# Education Percentage
# 0: no education: analfabet o estudis primaris incomplets
# 1: primary, 2: middle - primera etapa educació secundària
# 3: secondary - segona etapa educació secundària 
# 4:higher - FP, graus universitaris, masters, doctorat

construct_df['Level of Studies'] = ((educ_index_df['No education or Incomplete Primary School in people older than 15 (%)']*0)/100 + (educ_index_df['Primary School in people older than 15 (%)']*1)/100 + (educ_index_df['Middle Studies in people older than 15 (%)']*2)/100 + (educ_index_df['Secondary School in people older than 15 (%)']*3)/100 + (educ_index_df['Higher Studies in people older than 15 (%)']*4)/100)
construct_df['Students per Centre'] = educ_index_df['Students per Centre'].round().astype(int)
construct_df['Graduation Rate in 4th Grade'] = educ_index_df['Graduation Rate in 4th Grade (%)']/100
display(construct_df)
```

```python
# Apply minmax to level of studies
pop_col_studies = 'Level of Studies'
pop_col_students = 'Students per Centre'

# Reshape the data since the scaler expects a 2D array
construct_df[[pop_col_studies]] = scaler.fit_transform(construct_df[[pop_col_studies]])
construct_df[[pop_col_students]] = scaler.fit_transform(construct_df[[pop_col_students]])

display(construct_df)
```

```python
construct_df['Population Education'] = (construct_df['Level of Studies'] + construct_df['Students per Centre'] + construct_df['Graduation Rate in 4th Grade'])/3
construct_df.drop(columns={'Level of Studies', 'Students per Centre', 'Graduation Rate in 4th Grade'}, inplace=True)
display(construct_df)
```

```python
construct_df['Index'] = (construct_df['Accessibility'] + construct_df['Offer Quality'] + construct_df['Population Education'])/3
construct_df['Neighborhood Name'] = educ_index_df['Neighborhood Name']
construct_df.index = construct_df.index + 1
display(construct_df)
```

```python
#print(construct_df.max())
#print(construct_df.idxmax())
```

```python
# Correct Format for Tableau
construct_df['Accessibility'] = construct_df['Accessibility'] * 1000
construct_df['Offer Quality'] = construct_df['Offer Quality'] * 1000
construct_df['Population Education'] = construct_df['Population Education'] * 1000
construct_df['Index'] = construct_df['Index'] * 1000

construct_df['Accessibility'] = construct_df['Accessibility'].round().astype(int)
construct_df['Offer Quality'] = construct_df['Offer Quality'].round().astype(int)
construct_df['Population Education'] = construct_df['Population Education'].round().astype(int)
construct_df['Index'] = construct_df['Index'].round().astype(int)

display(construct_df)
```
