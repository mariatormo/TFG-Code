### ACCESSIBILITY

```python
# Accessibility
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
access_col = 'Number of Services'

# Reshape the data since the scaler expects a 2D array
health_idx_info_df[[access_col]] = scaler.fit_transform(health_idx_info_df[[access_col]])

display(health_idx_info_df)
```

```python
construct_df = pd.DataFrame(health_idx_info_df['Number of Services'])
construct_df.rename(columns={'Number of Services': 'Accessibility'}, inplace=True)
display(construct_df)
```

### OFFER QUALITY

```python
health_idx_info_df.head(5)
```

```python
# Count distinct values of centers type string (separated by comma)
def count_distinct_values(row): 
    if row == '':
        return 0
    values = row.split(',')
    distinct_values = set(values)
    return len(distinct_values)

# Apply to Centers Type column
construct_df['Count Centers Type'] = health_idx_info_df['Centers Type'].apply(count_distinct_values)
```

```python
# Apply Min-Max

# Offer Quality
offer_col = 'Count Centers Type'

# Reshape the data since the scaler expects a 2D array
construct_df[[offer_col]] = scaler.fit_transform(construct_df[[offer_col]])
construct_df.rename(columns={'Count Centers Type': 'Offer Quality'}, inplace=True)
display(construct_df)
```
